```python
links=[]
title=[]
dura=[]
def utubesearch(keyword: str, search_result: int):
  s=Search(keyword, limit=search_result)
  for i in range(len(s.result()['result'])):
    if ':' in s.result()['result'][i]['duration']:
      dura=[int(float(s.result()['result'][i]['duration'].replace(':', '')))]
    else:
      dura=[int(float(s.result()['result'][i]['duration'].replace('.', '')))]
      for k in dura:
        if k< 1200:
          links.append(s.result()['result'][i]['link'])
          title.append(s.result()['result'][i]['title'])
  return links, title
    
    
    
```

```python
def vidtoaud(video_path, keyword):
  file_list = [file for file in os.listdir(video_path) if os.path.splitext(file)[1] == '.3gpp']
  for i in range(len(file_list)):
    clip = mp.VideoFileClip(video_path+'/'+file_list[i]) 
    clip.audio.write_audiofile('/content/drive/MyDrive/Youtube Project/audio/'+keyword+str(i)+".mp3")
  return None
```

```python
import glob
import whisper

list_text=[]
for path in glob.glob('/content/drive/MyDrive/Youtube Project/audio/*.mp3'):
  model = whisper.load_model("base")
  result = model.transcribe(path)   
  list_text.append(result['text'])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel
from nltk import word_tokenize          
from nltk.stem import WordNetLemmatizer
import nltk
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
```

```python
class LemmaTokenizer:
    ignore_tokens = [',', '.', ';', ':', '"', '``', "''", '`']
    def __init__(self):
        self.wnl = WordNetLemmatizer()
    def __call__(self, doc):
        return [self.wnl.lemmatize(t) for t in word_tokenize(doc) if t not in self.ignore_tokens]

# Lemmatize the stop words
tokenizer=LemmaTokenizer()
token_stop = tokenizer(' '.join(stop_words))
```

```python

search_terms = keyword

vectorizer = TfidfVectorizer(stop_words=token_stop, 
                              tokenizer=tokenizer)
doc_vectors = vectorizer.fit_transform([search_terms] + list_text)

# Calculate similarity
cosine_similarities = linear_kernel(doc_vectors[0:1], doc_vectors).flatten()
document_scores = [item.item() for item in cosine_similarities[1:]]
document_scores
```

```python
df['document_similarity']=pd.Series(document_scores)
```
